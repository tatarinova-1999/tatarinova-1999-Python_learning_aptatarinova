#### Реализовать прототип консольной программы - проводника, для работы с файлами. Создать функции создания, удаления, перемещения, копирования(файла, папки) с использованием системы контроля версий git. Зарегистрироваться на Github и выгрузить с помощью git программу в созданный репозиторий. Прикрепить ссылку на репозиторий

```python
import os
import shutil

#Создание файла
def create_file():
    file_path = input('Введите путь (C:\\Users\\test\\test.py): ')
    open(file_path, 'w').close()
    print('Операция создания файла выполнена успешно')
#Создание папки 
def create_folder():
    folder_path = input('Введите путь (C:\\Users\\test): ')
    os.makedirs(folder_path)
    print('Операция создания папки выполнена успешно')
#Удаление файла/папки
def delete_item():
    path = input('Введите путь (C:\\Users\\test\\test.py):')
    if os.path.isfile(path):
        os.remove(path)
        print('Операция удаления файла выполнена успешно')
    elif os.path.isdir(path):
        shutil.rmtree(path)
        print('Операция удаления папки выполнена успешно')
#Перемещение 
def move_item():
    source = input('Введите исходный путь:')
    destination = input('Введите путь для перемещения:')
    shutil.move(source, destination)
    print('Операция перемещения выполнена успешно')
#Копирование файла/папки
def copy_item():
    source = input('Введите исходный путь:')
    destination = input('Введите путь для копирования:')
    if os.path.isfile(source):
        shutil.copy2(source, destination)
        print('Операция копирования файла выполнена успешно')
    elif os.path.isdir(source):
        shutil.copytree(source, destination)
        print('Операция копирования папки выполнена успешно')

def menu():
    try:
        while True:
            command = int(input('Выберите операцию: \n0. Выход из программы \n1. Создание \n2. Удаление \n3. Перемещение \n4. Копирование\n'))
            if command == 0:
                break
            elif command == 1: 
                com = int(input('Выберите объект: \n1. Файл \n2. Папка \n'))
                if com == 1:
                    create_file()
                elif com == 2:
                    create_folder()
            elif command == 2:
                delete_item()      
            elif command == 3:
                move_item()
            elif command == 4:
                copy_item()
            else: 
                print(f'Операция не найдена')
    except Exception as e:
        print(f'Ошибка!: {e}')

menu()
```
